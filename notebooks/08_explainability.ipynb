# 08 — Explainability (SHAP + LIME)

**Goal**: Generate global and local explanations for the best supervised model.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
from src.config_manager import ConfigurationManager
from src.constants import CLEAN_DATA_FILE, BEST_CLASSIFIER_FILE
from src.utils import load_object
from src.explainability.shap_explainer import SHAPExplainer
from src.explainability.lime_explainer import LIMEExplainer

cm = ConfigurationManager()
df = pd.read_csv(CLEAN_DATA_FILE)
df.columns = [c.replace(' ', '_') for c in df.columns]
X = df.drop(columns=['cluster'])

model = load_object(BEST_CLASSIFIER_FILE)
shap_exp = SHAPExplainer(cm)
shap_artifact = shap_exp.explain(model, X)
```

## Global Feature Importance (SHAP)

```python
import matplotlib.pyplot as plt
import shap
if shap_artifact.shap_values is not None:
    shap.summary_plot(shap_artifact.shap_values, shap_artifact.sample, show=False)
    plt.show()
```

## LIME Local Explanation

```python
lime_exp = LIMEExplainer(cm)
lime_artifact = lime_exp.explain(model, X, y=df['cluster'])
for exp in lime_artifact.sample_explanations[:3]:
    print(f'--- Instance {exp["index"]} (class {exp["predicted_class"]}) ---')
    for feat, weight in exp['as_list'][:5]:
        print(f'  {feat:50s} {weight:+.4f}')
```
