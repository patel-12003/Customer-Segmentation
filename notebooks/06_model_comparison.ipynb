# 06 — Supervised Model Comparison

**Goal**: Train and compare 11 supervised classifiers on the cluster labels.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
from src.config_manager import ConfigurationManager
from src.constants import CLEAN_DATA_FILE
from src.models.supervised import SupervisedModelTrainer
from src.features.selection import FeatureSelector

cm = ConfigurationManager()
df = pd.read_csv(CLEAN_DATA_FILE)
df.columns = [c.replace(' ', '_') for c in df.columns]

# Select features first
selector = FeatureSelector(cm)
sel = selector.initiate_feature_selection(df, target_column='cluster')

trainer = SupervisedModelTrainer(cm)
artifact = trainer.run(df, sel.selected_features)
```

## Leaderboard

```python
rows = [{'name': r.name, 'accuracy': r.accuracy, 'f1': r.f1, 'cv_mean': r.cv_mean_accuracy, 'cv_std': r.cv_std_accuracy} for r in artifact.results]
pd.DataFrame(rows).sort_values('cv_mean', ascending=False)
```

## Best Model

```python
print(f'Best: {artifact.best_model_name} (acc={artifact.best_model_result.accuracy:.4f}, CV={artifact.best_model_result.cv_mean_accuracy:.4f})')
```

## Confusion Matrix

```python
import matplotlib.pyplot as plt
from src.visualization.plots import Plotter
fig = Plotter.confusion_matrix_heatmap(artifact.best_model_result.confusion_matrix)
plt.show()
```
