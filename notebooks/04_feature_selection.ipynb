# 04 — Feature Selection

**Goal**: Rank features by Mutual Information with the cluster label and select the top-K.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
from src.config_manager import ConfigurationManager
from src.constants import CLEAN_DATA_FILE
from src.features.selection import FeatureSelector

cm = ConfigurationManager()
df = pd.read_csv(CLEAN_DATA_FILE)
df.columns = [c.replace(' ', '_') for c in df.columns]

selector = FeatureSelector(cm)
artifact = selector.initiate_feature_selection(df, target_column='cluster')
print('Selected:', artifact.selected_features)
```

## Feature Ranking

```python
ranking = sorted(artifact.scores.items(), key=lambda kv: kv[1], reverse=True)
for i, (feat, score) in enumerate(ranking, 1):
    marker = '✅' if feat in artifact.selected_features else '  '
    print(f'{marker} {i:2d}. {feat:25s} {score:.4f}')
```

## Conclusion

Top 15 features selected by Mutual Information. The selector state is persisted to `saved_models/feature_selector.joblib`.
