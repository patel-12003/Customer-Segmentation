# 02 — Exploratory Data Analysis

**Goal**: Visualise distributions, correlations and segment-level differences.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.constants import CLEAN_DATA_FILE
from src.visualization.plots import apply_dark_theme, DARK_PALETTE
apply_dark_theme()

df = pd.read_csv(CLEAN_DATA_FILE)
df.columns = [c.replace(' ', '_') for c in df.columns]
df.head()
```

## 1. Cluster Size Distribution

```python
fig, ax = plt.subplots(figsize=(7,4), constrained_layout=True)
df['cluster'].value_counts().sort_index().plot.bar(ax=ax, color=DARK_PALETTE['palette'])
ax.set_title('Cluster Sizes')
plt.show()
```

## 2. Distributions by Cluster

```python
num_cols = ['Income','Total_Spending','Age','Recency','Wines','Meat']
fig, axes = plt.subplots(2, 3, figsize=(15,8), constrained_layout=True)
for ax, col in zip(axes.flat, num_cols):
    for i, c in enumerate(sorted(df['cluster'].unique())):
        sub = df[df['cluster']==c][col]
        sns.kdeplot(sub, ax=ax, label=f'Cluster {c}', color=DARK_PALETTE['palette'][i], fill=True, alpha=0.3)
    ax.set_title(col)
    ax.legend()
plt.show()
```

## 3. Correlation Heatmap

```python
fig, ax = plt.subplots(figsize=(11,9), constrained_layout=True)
sns.heatmap(df.drop(columns=['cluster']).corr(), cmap='coolwarm', center=0, ax=ax, linewidths=0.3)
ax.set_title('Feature Correlation')
plt.show()
```

## 4. Cluster Mean Profiles

```python
profile = df.groupby('cluster').mean().T
profile
```

## 5. Key Insights

* Cluster 1 (Premium High-Spenders) — highest Income, Wines, Meat, Catalog purchases.
* Cluster 0 (Budget-Conscious Families) — most Children, highest NumWebVisitsMonth.
* Cluster 2 (Mainstream Loyalists) — balanced across features, largest segment.
