# 05 — Clustering

**Goal**: Train, evaluate and compare 10 clustering algorithms.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
from src.config_manager import ConfigurationManager
from src.constants import CLEAN_DATA_FILE
from src.models.clustering import ClusteringEngine
from src.visualization.plots import apply_dark_theme
apply_dark_theme()

cm = ConfigurationManager()
df = pd.read_csv(CLEAN_DATA_FILE)
df.columns = [c.replace(' ', '_') for c in df.columns]
X = df.drop(columns=['cluster'])

engine = ClusteringEngine(cm)
artifact = engine.run(X)
```

## Leaderboard

```python
import pandas as pd
rows = [{'name': r.name, 'n_clusters': r.n_clusters, 'silhouette': r.silhouette,
         'davies_bouldin': r.davies_bouldin, 'calinski_harabasz': r.calinski_harabasz,
         'stability': r.stability, 'interpretability': r.interpretability} for r in artifact.results]
leaderboard = pd.DataFrame(rows).sort_values('silhouette', ascending=False)
leaderboard
```

## Best Model

```python
print(f'Best: {artifact.best_model_name} (silhouette={artifact.best_score:.4f})')
```
